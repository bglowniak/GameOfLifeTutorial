```python
def init_glider_world():
    world = np.zeros((10, 10), dtype=int)
    # hardcode glider pattern
    world[9,1] = ALIVE
    world[8,2] = ALIVE
    world[7,0] = ALIVE
    world[7,1] = ALIVE
    world[7,2] = ALIVE
    return world

world = init_glider_world()
plot_world(world)
```

```python
def plot_glider_series():
    world = init_glider_world()
    series = timeseries(world, 25)

    cmap = colors.ListedColormap(['white', 'red'])

    for i in range(0, 4):
        plt.subplot(2, 4, i + 1)
        plt.pcolor(series[i], cmap=cmap, edgecolor="black")
        plt.axis('square')
        plt.title("t = " + str(i))
        
        plt.subplot(2, 4, i + 5)
        plt.pcolor(series[(i + 1) * 5], cmap=cmap, edgecolor = "black")
        plt.axis('square')
        plt.title("t = " + str((i + 1) * 5))

    plt.show()
    
plot_glider_series()
```
