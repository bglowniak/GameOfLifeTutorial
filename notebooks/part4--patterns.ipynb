```python
from pattern_worlds import init_fixed_world

def plot_fixed_series():
    world = init_fixed_world()
    series = timeseries(world, 8)

    for i in range(0, 3):
        plt.subplot(1, 3, i + 1)
        plt.pcolor(series[i * 3], cmap=cmap, edgecolor="black")
        plt.axis('square')
        plt.title("t = " + str(i * 3))

    plt.show()
    
plot_fixed_series()
```

So far, the patterns we've investigated don't really mean much. They may be interesting to look at, but 

We now investigate behaviors that *move*, which as we will see, have some fairly profound implications for what you can do with Game of Life.

```python
from pattern_worlds import init_glider_world

world = init_glider_world()
plot_world(world)
```

```python
def plot_glider_series():
    world = init_glider_world()
    series = timeseries(world, 25)

    for i in range(0, 4):
        plt.subplot(2, 4, i + 1)
        plt.pcolor(series[i], cmap=cmap, edgecolor="black")
        plt.axis('square')
        plt.title("t = " + str(i))
        
        plt.subplot(2, 4, i + 5)
        plt.pcolor(series[(i + 1) * 5], cmap=cmap, edgecolor = "black")
        plt.axis('square')
        plt.title("t = " + str((i + 1) * 5))

    plt.show()
    
plot_glider_series()
```

Above, we plot timesteps 0-3 to demonstrate the period of a glider, then t= 5 - 20 in steps of 5 to demonstrate how a glider transmits along a diagonal across the board.
